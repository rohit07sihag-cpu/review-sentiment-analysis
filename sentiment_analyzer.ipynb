```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")

reviews = [
    "Battery backup is amazing, lasts 2 full days",
    "Camera quality is disappointing for this price range",
    # add 15-20 more real reviews here
]

for r in reviews:
    result = classifier(r)[0]
    print(f"{r[:50]}... -> {result['label']} ({result['score']:.2f})")
```

```python
reviews = [
    "Battery backup is amazing and lasts two days.",
    "Camera quality is disappointing.",
    "The display is bright and beautiful.",
    "Charging is very fast.",
    "The phone gets hot while gaming.",
    "Excellent value for money.",
    "Sound quality is average.",
    "Very smooth performance.",
    "Battery drains quickly.",
    "The build quality feels premium.",
    "The screen is sharp and colorful.",
    "Fingerprint sensor works perfectly.",
    "The camera performs poorly in low light.",
    "Software experience is excellent.",
    "The phone is a bit heavy.",
    "Call quality is crystal clear.",
    "Packaging was damaged.",
    "Storage is enough for daily use.",
    "Highly recommended.",
    "Not worth the price."
]
```

```python
for review in reviews:
    result = classifier(review)[0]

    print("Review:", review)
    print("Sentiment:", result["label"])
    print("Confidence:", round(result["score"],3))
    print("-"*50)
```

# Insights

The pretrained Hugging Face sentiment analysis model correctly classified positive and negative customer reviews.

Positive reviews mainly praised battery life, display, performance and value for money.

Negative reviews mainly complained about camera quality, heating issues and battery drain.
